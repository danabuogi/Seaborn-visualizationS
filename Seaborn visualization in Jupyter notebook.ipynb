# Notes
- In the above map, as you go further out, the less instances you have.
- below if you add fill is equal to True you can see where the highest concentration is.

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde",fill=True, color="navy")
```

# Note
- You can also add a cmap color as below

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde",fill=True, color="navy",cmap="YlGnBu")
```

# Note
- You can also change the kind to hex as shown below.
- However, you are to remove fill because a hex does not have fill.

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="hex", color="navy",cmap="YlGnBu")
```
